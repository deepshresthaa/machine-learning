```python
## counting
n_total=df.shape[0]

given_yes=df[df['play']=="Yes"]
given_no=df[df["play"]=='No']

n_of_yes=given_yes.shape[0]
n_of_no=given_no.shape[0]

p_yes=n_of_yes/n_total
p_no=n_of_no/ n_total
print(n_of_yes)
print(n_of_no)
print(p_yes)
print(p_no)
```

## probability calculation for 'outlook'

```python
n_of_overcast_when_yes=given_yes[given_yes['outlook']=="Overcast"].shape[0]
n_of_sunny_when_yes=given_yes[given_yes['outlook']=="Sunny"].shape[0]
n_of_rain_when_yes=given_yes[given_yes['outlook']=="Rain"].shape[0]


p_outlook_overcast_given_yes= n_of_overcast_when_yes / n_of_yes# P(outlook=overcase | yes)
p_outlook_rain_given_yes= n_of_rain_when_yes / n_of_yes# P(outlook=rain | yes)
p_outlook_sunny_given_yes= n_of_sunny_when_yes / n_of_yes# P(outlook=sunny | yes)
p_yes=n_of_yes/ n_total


## outlook when play='No'

n_of_overcast_when_no=given_no[given_no['outlook']=="Overcast"].shape[0]
n_of_sunny_when_no=given_no[given_no['outlook']=="Sunny"].shape[0]
n_of_rain_when_no=given_no[given_no['outlook']=="Rain"].shape[0]


p_outlook_overcast_given_no= n_of_overcast_when_no / n_of_no# P(outlook=overcase | yes)
p_outlook_rain_given_no= n_of_rain_when_no / n_of_no# P(outlook=rain | yes)
p_outlook_sunny_given_no= n_of_sunny_when_no / n_of_no# P(outlook=sunny | yes)
p_yes=n_of_no/ n_total

print(p_outlook_overcast_given_yes)
print(p_outlook_rain_given_yes)
print(p_outlook_sunny_given_yes)
print(p_outlook_overcast_given_no)
print(p_outlook_rain_given_no)
print(p_outlook_sunny_given_no)
```

```python
print(p_yes)
```

## probability calculation for `temp`

```python
given_yes
```

```python
n_of_hot_when_yes=given_yes[given_yes["temp"]=='Hot'].shape[0]
n_of_mild_when_yes=given_yes[given_yes['temp']=="Mild"].shape[0]
n_of_cool_when_yes=given_yes[given_yes["temp"]=="Cool"].shape[0]

p_of_temp_hot_when_yes=n_of_hot_when_yes / n_of_yes # p(temp="Hot" | yes)
p_of_temp_mild_when_yes=n_of_mild_when_yes/ n_of_yes
p_of_temp_cool_when_yes=n_of_cool_when_yes / n_of_yes



n_of_hot_when_no=given_no[given_no["temp"]=='Hot'].shape[0]
n_of_mild_when_no=given_no[given_no['temp']=="Mild"].shape[0]
n_of_cool_when_no=given_no[given_no["temp"]=="Cool"].shape[0]

p_of_temp_hot_when_no=n_of_hot_when_no / n_of_no # p(temp="Hot" | yes)
p_of_temp_mild_when_no=n_of_mild_when_no/ n_of_no
p_of_temp_cool_when_no=n_of_cool_when_no / n_of_no

print(p_of_temp_hot_when_yes)
print(p_of_temp_mild_when_yes)
print(p_of_temp_cool_when_yes)

print(p_of_temp_hot_when_no)
print(p_of_temp_mild_when_no)
print(p_of_temp_cool_when_no)
```

## humidity when play="Yes"

```python
n_of_normal_when_yes=given_yes[given_yes["humidity"]=="Normal"].shape[0]
n_of_high_when_yes=given_yes[given_yes["humidity"]=="High"].shape[0]
p_of_humidity_normal_when_yes=n_of_normal_when_yes/ n_of_yes # p(humidity=Normal | Yes)
p_of_humidity_high_when_yes=n_of_high_when_yes / n_of_yes

n_of_normal_when_no=given_no[given_no["humidity"]=="Normal"].shape[0]
n_of_high_when_no=given_no[given_no["humidity"]=="High"].shape[0]
p_of_humidity_normal_when_no=n_of_normal_when_no/ n_of_no # p(humidity=Normal | No)
p_of_humidity_high_when_no=n_of_high_when_no / n_of_no

print(p_of_humidity_normal_when_yes)
print(p_of_humidity_high_when_yes)
print(p_of_humidity_normal_when_no)
print(p_of_humidity_high_when_no)
```

## wind when play="Yes"

```python
given_yes["wind"]
```

```python
n_of_weak_when_yes=given_yes[given_yes["wind"]=="Weak"].shape[0]
n_of_strong_when_yes=given_yes[given_yes["wind"]=="Strong"].shape[0]
p_of_wind_weak_when_yes=n_of_weak_when_yes / n_of_yes
p_of_wind_strong_when_yes= n_of_strong_when_yes /n_of_yes


n_of_weak_when_no=given_no[given_no["wind"]=="Weak"].shape[0]
n_of_strong_when_no=given_no[given_no["wind"]=="Strong"].shape[0]
p_of_wind_weak_when_no=n_of_weak_when_no / n_of_no
p_of_wind_strong_when_no= n_of_strong_when_no /n_of_no

print(p_of_wind_weak_when_yes)
print(p_of_wind_strong_when_yes)
print(p_of_wind_weak_when_no)
print(p_of_wind_strong_when_no)

```

```python
# p(yes | sunny, mild, normal, strong)= p(sunny, mild, normal , strong | yes)*p(yes)
# P(outlook=sunny|yes)*p(temp=mild|yes)*p(humidity=normal|yes)*p(wind=strong |yes)
posterior_yes=p_yes * p_outlook_sunny_given_yes * p_of_temp_mild_when_yes * p_of_humidity_normal_when_yes * p_of_wind_strong_when_yes
posterior_no=p_no * p_outlook_sunny_given_no * p_of_temp_mild_when_no * p_of_humidity_normal_when_no * p_of_wind_strong_when_no
```

```python
print(posterior_yes)
print(posterior_no)
```

### since, `posterior_yes > posterior_no`, so `play="yes"` is predicted
